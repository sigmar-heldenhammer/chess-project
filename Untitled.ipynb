```python
from pathlib import Path
import chess
import chess.pgn
import matplotlib.pyplot as plt
```

```python


def plot_pgn_mobility_and_piececount(pgn_path: str, color: chess.Color):
    """
    Inputs:
      - pgn_path: path to a PGN file
      - color: chess.WHITE or chess.BLACK

    Outputs:
      - Displays two line charts:
          1) Move (ply) vs legal moves available to `color`
          2) Move (ply) vs total pieces on board
    Notes:
      - Uses the *first* game in the PGN and its mainline.
      - X-axis is ply number (1 = after White's 1st move, 2 = after Black's reply, etc.).
    """
    pgn_path = Path(pgn_path)
    if not pgn_path.exists():
        raise FileNotFoundError(f"PGN not found: {pgn_path}")

    with pgn_path.open("r", encoding="utf-8", errors="replace") as f:
        game = chess.pgn.read_game(f)

    if game is None:
        raise ValueError(f"No game found in PGN: {pgn_path}")

    board = game.board()

    ply = []
    legal_moves_for_color = []
    piece_counts = []

    def count_legal_moves_for(board: chess.Board, side: chess.Color) -> int:
        # python-chess only generates legal moves for the side to move, so temporarily flip turn.
        tmp = board.copy(stack=False)
        tmp.turn = side
        return tmp.legal_moves.count()

    # Record state after each ply in the mainline
    ply_idx = 0
    for move in game.mainline_moves():
        board.push(move)
        ply_idx += 1

        ply.append(ply_idx)
        legal_moves_for_color.append(count_legal_moves_for(board, color))
        piece_counts.append(len(board.piece_map()))

    # Plot 1: legal moves for color
    plt.figure()
    plt.plot(ply, legal_moves_for_color)
    plt.xlabel("Move (ply)")
    plt.ylabel(f"Legal moves for {'White' if color == chess.WHITE else 'Black'}")
    plt.title("Legal moves by ply")
    plt.show()

    # Plot 2: total pieces on board
    plt.figure()
    plt.plot(ply, piece_counts)
    plt.xlabel("Move (ply)")
    plt.ylabel("Total pieces on board")
    plt.title("Piece count by ply")
    plt.show()
```

```python
plot_pgn_mobility_and_piececount("match.pgn", chess.WHITE)
```

```python
plot_pgn_mobility_and_piececount("match.pgn", chess.BLACK)
```
